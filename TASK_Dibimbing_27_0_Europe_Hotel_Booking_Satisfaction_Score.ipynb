```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import data
df = pd.read_csv('Europe Hotel Booking Satisfaction Score.csv')
df
```

```python
# Lihat sekilas data
print(df.head())
```

```python
print(df.info())
```

```python
# Tampilkan nama kolom untuk memastikan kolom yang tersedia
print(df.columns)
```

```python
# Statistik deskriptif
print(df.describe())
```

```python
# Memeriksa distribusi nilai untuk kolom penting
sns.histplot(df['satisfaction'], kde=True)
plt.title('Distribusi Kepuasan Pelanggan Hotel')
plt.xlabel('Kepuasan')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Pilih hanya kolom numerik
numeric_df = df.select_dtypes(include=['float64', 'int64'])
```

```python
# Hitung korelasi
corr = numeric_df.corr()
```

```python
# Plot heatmap korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap Korelasi')
plt.show()
```

```python
# Boxplot untuk melihat hubungan antara kepuasan pelanggan dan beberapa fitur

# Kepuasan Pelanggan berdasarkan Tipe Booking
if 'Type Of Booking' in df.columns and 'satisfaction' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Type Of Booking', y='satisfaction', data=df)
    plt.title('Kepuasan Pelanggan berdasarkan Tipe Booking')
    plt.show()
else:
    print("Kolom 'Type Of Booking' atau 'satisfaction' tidak ditemukan dalam DataFrame")
```
